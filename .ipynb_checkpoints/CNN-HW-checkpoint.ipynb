## 插入套件、資料庫

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape
```

## 設定標籤

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

## 因為設定灰階，所以只有一個channel，接下來轉變資料格式

```python
x_train = x_train.reshape(60000, 28, 28, 1) / 255
```

```python
x_test = x_test.reshape(10000, 28, 28, 1) / 255
```

```python
x_train[87].shape
```

```python
y_train[87]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

## 讀入函數

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
```

## 打造函數學習機
### 改變filter數量

```python
model = Sequential()
```

```python
model.add(Conv2D(32, (3,3), padding='same',
                input_shape=(28,28,1),
                activation='relu'))   #過濾器3x3, 32個
```

```python
model.add(MaxPooling2D(pool_size=(2,2))) #(14,14,32)
```

```python
model.add(Conv2D(64, (3,3), padding='same',
                activation='relu'))    #(14,14,64)
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))  #(7, 7, 64)
```

```python
model.add(Conv2D(128, (3,3), padding='same',
                activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())  #拉平
```

```python
model.add(Dense(54, activation='relu'))
```

```python
model.add(Dense(10, activation='softmax'))
```

## 檢視神經網路

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer=SGD(lr=0.087),
             metrics=['accuracy'])
```

## 改變batchsize, epoch

```python
model.fit(x_train, y_train, batch_size=200, epochs=10)
```

## 預測 

```python
result = model.predict_classes(x_test)
```

```python
def my_predict(n):
    print('我的 CNN 預測是', result[n])
    X = x_test[n].reshape(28,28)
    plt.imshow(X, cmap='Greys')
```

```python
from ipywidgets import interact_manual
```

```python
interact_manual(my_predict, n=(0, 9999))
```

```python
score = model.evaluate(x_test, y_test)
```

```python
loss, acc = score
```

```python
print('測試資料的正確率為', acc)
```

```python
model.save('myCNNmodel.h5')
```
